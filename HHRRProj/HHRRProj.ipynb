```python
import pandas as pd
import itertools

def corrank(df):
    # calcular correlaciones numéricas
    df_corr = df.corr(numeric_only=True)

    # generar pares de columnas y sus correlaciones
    df_corr_pairs = pd.DataFrame(
        [[(i, j), df_corr.loc[i, j]] for i, j in itertools.combinations(df_corr.columns, 2)],
        columns=['pairs', 'corr']
    )

    # devolver el dataframe ordenado
    return df_corr_pairs.sort_values(by='corr', ascending=False).reset_index(drop=True)

# usar la función
dfrk = corrank(df)

# mostrar los 25 pares más correlacionados
print(dfrk.head(25))
```

#### Prueba T para la Diferencia de Medias en Satisfacción Laboral (Deserción vs. No Deserción)

Prueba T de una muestra (Midiendo el Nivel de Satisfacción)
Una prueba t de una muestra verifica si la media de una muestra difiere de la media de la población. Probemos si el nivel promedio de satisfacción de los empleados que tuvieron deserción difiere de la población total de empleados.

* Prueba de Hipótesis: ¿Existe una diferencia significativa en las medias del nivel de satisfacción entre los empleados que desertaron y la población total de empleados?

* Hipótesis Nula: La hipótesis nula sería que no hay diferencia en el nivel de satisfacción entre los empleados que desertaron y la población total de empleados.

* Hipótesis Alternativa: La hipótesis alternativa sería que existe una diferencia en el nivel de satisfacción entre los empleados que desertaron y la población total de empleados.

```python
df['Attrition'] = df['Attrition'].map({'No':0,'Yes':1})
```

```python
# Comparemos las medias del total de empleados con los que se fueron de la organizacion
indiceSatisfaccionTotal= df['JobSatisfaction']
indiceSatisfaccionDesvinculados=df[df['Attrition']==1]['JobSatisfaction']


print( 'La media poblacional es: ' + str(indiceSatisfaccionTotal.mean()) )
print( 'La media de los empleados que desvincularon: ' + str(indiceSatisfaccionDesvinculados.mean()) )
```

* Realicemos una prueba t con un nivel de confianza del 95% para ver si rechaza correctamente la hipótesis nula de que la muestra proviene de la misma distribución que la población de empleados. Para realizar una prueba t de una muestra, podemos usar la función stats.ttest_1samp().

```python
import scipy.stats as stats

# Realizar la prueba t de una muestra
resultado = stats.ttest_1samp(
    a=indiceSatisfaccionDesvinculados,  # muestra
    popmean=indiceSatisfaccionTotal.mean()  # media poblacional
)

# Extraer valores
t_stat = resultado.statistic
p_val = resultado.pvalue
alpha = 0.05  # nivel de significancia 5%

# Imprimir resultados de forma clara
print("=== One-Sample T-Test ===")
print(f"Media de la población: {indiceSatisfaccionTotal.mean():.3f}")
print(f"Media de la muestra: {indiceSatisfaccionDesvinculados.mean():.3f}")
print(f"Estadístico t: {t_stat:.3f}")
print(f"P-valor: {p_val:.6f}")

# Conclusión
if p_val < alpha:
    print(f"Resultado: Rechazamos la hipótesis nula al nivel de {alpha*100:.0f}% de significancia.")
    print("Interpretación: La satisfacción de los empleados desvinculados es significativamente distinta de la media poblacional.")
else:
    print(f"Resultado: No rechazamos la hipótesis nula al nivel de {alpha*100:.0f}% de significancia.")
    print("Interpretación: No hay evidencia de que la satisfacción de los empleados desvinculados difiera de la media poblacional.")
```

#### Conclusión del One-Sample T-Test

Se realizó una prueba T de una muestra para evaluar si el nivel de satisfacción de los empleados que se desvincularon (*Attrition = Yes*) difiere significativamente del promedio de toda la población de empleados.

- **Media de satisfacción de la población:** 2.73  
- **Media de satisfacción de empleados desvinculados:** 2.47  
- **Estadístico t:** 3.58  
- **P-valor:** 0.00041  

Dado que el p-valor es menor al nivel de significancia del 5% (0.05), se **rechaza la hipótesis nula**. Esto indica que existe una diferencia estadísticamente significativa entre la satisfacción de los empleados que se fueron y la media poblacional.

⚠️ **Nota:** Aunque la diferencia es estadísticamente significativa, su relevancia práctica podría ser pequeña. Se recomienda complementar con más análisis o datos adicionales para evaluar el impacto real sobre la organización.

### Validacion Cruzada para identificar el mejor modelo para prediccion

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import numpy as np

# 1. Separar target y features
y = df["Attrition"].copy()
X = df.drop("Attrition", axis=1)

# 2. Split en train/test (manteniendo proporción de clases)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=7
)

# 3. Escalado con MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Definir modelos
models = {
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(eval_metric="logloss")
}

# 5. Evaluación con cross-validation en TRAIN
for name, model in models.items():
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring="f1")
    print(f"{name} - F1 promedio (sin SMOTE): {np.mean(scores):.3f}")
```

### XGBoost Entrenamiento + Evaluación
1. Calcular `scale_pos_weight` por balance de clases.  
2. Entrenar modelo con hiperparámetros fijos.  
3. Validar con ROC-AUC (CV=5).  
4. Predecir probabilidades en test.  
5. Ajustar threshold óptimo según F1.  
6. Evaluar con reportes, ROC-AUC y matriz de confusión.  

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve
from sklearn.model_selection import StratifiedKFold, cross_val_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# ============================
# 1️⃣ Calcular scale_pos_weight
# ============================
neg, pos = (y_train == 0).sum(), (y_train == 1).sum()
scale = neg / pos
print("scale_pos_weight =", scale)

# ============================
# 2️⃣ Definir y entrenar modelo
# ============================
xgb = XGBClassifier(
    learning_rate=0.01,
    n_estimators=500,
    max_depth=5,
    min_child_weight=5,
    gamma=0.1,
    subsample=0.6,
    colsample_bytree=1.0,
    scale_pos_weight=scale,
    eval_metric="logloss",
    random_state=7
)
xgb.fit(X_train, y_train)

# ============================
# 3️⃣ Validación: ROC-AUC CV
# ============================
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
roc_auc_scores = cross_val_score(xgb, X_train, y_train, cv=skf, scoring='roc_auc')
print("ROC-AUC cross-validation mean:", roc_auc_scores.mean())

# ============================
# 4️⃣ Predicciones
# ============================
y_prob = xgb.predict_proba(X_test)[:, 1]

# ============================
# 5️⃣ Threshold tuning basado en F1
# ============================
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
f1_scores = 2*precision*recall/(precision+recall)
best_thresh = thresholds[np.argmax(f1_scores)]
print("Best threshold based on F1:", best_thresh)

y_pred = (y_prob >= best_thresh).astype(int)

# ============================
# 6️⃣ Evaluación final
# ============================
print("=== XGBoost con threshold optimizado ===")
print(classification_report(y_test, y_pred, digits=3))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
plt.title("Confusion Matrix - XGBoost")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

1. Crear `TreeExplainer` y calcular `shap_values`.  
2. Mostrar importancia global (summary plot + bar plot).  
3. Analizar variable específica (`OverTime`) y su dirección de impacto.  
4. Generar `force plot` para explicar un individuo en detalle.  

```python
import shap
import pandas as pd
import matplotlib.pyplot as plt

# Explicador basado en XGBoost
explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X_test)

# 1. Importancia global (summary plot)
shap.summary_plot(shap_values, X_test)

# 2. Efecto individual de una variable
shap.summary_plot(shap_values, X_test, plot_type="bar")

# Create shap_df DataFrame
shap_df = pd.DataFrame(shap_values, columns=X_test.columns)
shap_df['OverTime'] = X_test['OverTime'].values # Add the original OverTime values

# Conteo de casos por valor de OverTime y su signo de SHAP
shap_df["impact"] = shap_df["OverTime"].map(lambda x: "High" if x==1 else "Low")
shap_df["direction"] = shap_df["OverTime"].map(lambda x: "Push_Retain" if x==0 else "Push_Attrition")

print(shap_df.groupby(["impact","direction"]).size())
```

### Interpretación de los SHAP Values

Los SHAP (SHapley Additive exPlanations) values nos ayudan a entender la contribución de cada característica individual a la predicción de nuestro modelo para cada instancia. En esencia, explican por qué el modelo hizo una predicción particular para un empleado específico, mostrando cómo cada característica empuja la predicción desde el valor base (la predicción promedio) hacia el valor de predicción final.

#### SHAP Summary Plot (Gráfico Resumen de SHAP)

El gráfico resumen de SHAP nos da una visión general de la importancia global de las características y su impacto en las predicciones del modelo.

- **Eje Y:** Muestra las características ordenadas por su importancia global (basada en la magnitud promedio de los SHAP values). Las características en la parte superior son las más influyentes.
- **Eje X:** Representa el valor SHAP de la característica. Este valor indica cuánto contribuye la característica a la predicción de deserción para una instancia particular. Un valor SHAP positivo empuja la predicción hacia la deserción (clase 1), mientras que un valor SHAP negativo empuja la predicción hacia la no deserción (clase 0).
- **Color:** El color de cada punto representa el valor de la característica para esa instancia. Típicamente, el rojo indica valores altos de la característica y el azul indica valores bajos (aunque esto puede variar dependiendo de la característica).
- **Dispersión de Puntos:** Cada punto en la fila de una característica representa un empleado en el conjunto de prueba. La dispersión horizontal de los puntos muestra el rango de impacto de esa característica en las predicciones.

**Interpretación del gráfico resumen:**

- **OverTime (Horas Extra):** Es la característica más importante. Los puntos rojos (alto valor de OverTime, es decir, Yes) tienen SHAP values positivos significativos, lo que significa que trabajar horas extra **aumenta fuertemente la probabilidad de que un empleado deserte** según el modelo. Los puntos azules (bajo valor de OverTime, es decir, No) tienen SHAP values negativos, lo que indica que no trabajar horas extra **disminuye la probabilidad de deserción.**
- **MonthlyIncome (Ingreso Mensual):** La segunda característica más importante. Los puntos azules (bajo ingreso) tienen SHAP values positivos, lo que sugiere que un ingreso bajo **aumenta la probabilidad de deserción**. Los puntos rojos (alto ingreso) tienen SHAP values negativos, lo que indica que un ingreso alto **disminuye la probabilidad de deserción**.
- **StockOptionLevel (Nivel de Opción de Acciones):** Los puntos azules (bajo nivel de opción de acciones) tienen SHAP values positivos, lo que implica que un nivel bajo de opciones de acciones **incrementa la probabilidad de deserción**. Los puntos rojos (alto nivel de opción de acciones) tienen SHAP values negativos, **reduciendo la probabilidad de deserción**.
- **Distance From Home (Distancia Desde Casa):** Parece que vivir más lejos (puntos rojos/altos valores) tiende a tener SHAP values positivos, **aumentando ligeramente la probabilidad de deserción**.
- **NumCompaniesWorked (Número de Empresas Trabajadas):** Tener un alto número de empresas trabajadas (puntos rojos) parece tener SHAP values positivos, **incrementando la probabilidad de deserción**.
- **Age (Edad):** Los puntos azules (baja edad/jóvenes) tienen SHAP values positivos, lo que sugiere que ser más joven **aumenta la probabilidad de deserción**. Los puntos rojos (alta edad/mayores) tienen SHAP values negativos, **disminuyendo la probabilidad de deserción**.

En general, el gráfico resume visualmente las relaciones identificadas en el EDA y la importancia que el modelo XGBoost les asigna para predecir la deserción.

#### SHAP Dependence Plot (Gráfico de Dependencia de SHAP) - Ejemplo con MonthlyIncome

El gráfico de dependencia muestra la relación entre el valor de una característica específica y su SHAP value correspondiente para cada instancia. Ayuda a visualizar cómo el impacto de una característica cambia a medida que su valor varía.

- **Eje X:** Muestra el valor real de la característica seleccionada (en este caso, MonthlyIncome).
- **Eje Y:** Muestra el valor SHAP para esa característica.
- **Color:** El color de los puntos puede representar el valor de otra característica con la que la característica seleccionada podría interactuar (por defecto, a menudo muestra la característica con la interacción más fuerte).

**Interpretación del gráfico de dependencia para MonthlyIncome:**

Observamos que para valores bajos de `MonthlyIncome` (aproximadamente por debajo de 5000-7000), los puntos tienen SHAP values mayormente positivos, lo que indica que tener un ingreso bajo contribuye positivamente a la predicción de deserción. A medida que el `MonthlyIncome` aumenta, los SHAP values tienden a volverse más negativos, lo que significa que un ingreso más alto reduce la probabilidad predicha de deserción. Esto confirma la fuerte relación inversa entre el ingreso mensual y la deserción identificada en el EDA y validada por el modelo.
